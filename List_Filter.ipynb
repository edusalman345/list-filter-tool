```python
def valueFinder(user_list):
    filteredList = []
    value = int(input("Enter the value: "))

    for x in user_list:
        if x > value:
            filteredList.append(x)

    print(filteredList)

def evenFinder(user_list):
  filteredList = []
  for x in user_list:
    if x %2 ==0:
      filteredList.append(x)
  print(filteredList)


def oddFinder(user_list):
  filteredList = []
  for x in user_list:
    if x %2 !=0:
      filteredList.append(x)
  print(filteredList)


def stringLengthFinder(user_list):
  filteredList = []
  value = int(input("Enter the length: "))

  for item in user_list:
      length = 0
      item = str(item)
      for x in item:
        length += 1
      if length >= value:
        filteredList.append(item)

  print(filteredList)



def empty(user_list):
  filteredList = []
  for x in user_list:
    if x != '':
      filteredList.append(x)
  print(filteredList)
```

```python

user_input = input("Enter items sepeatated by space: ")
if user_input == '':
  print("Please enter something")
else:
  user_list = user_input.split()

```

```python

user_list = [int(x) for x in user_list]
```

```python
options = ['Find value Greater then a certain number.','Find Even Numbers', 'Find Odd Numbers.','Find String longer than certain length.','Remove Empty Values.']

for index,op in enumerate(options, start = 1):
  print(f"{index})", op)
```

```python
option = input("Enter the option number here: ")
```

```python
match option:
  case '1':
    valueFinder(user_list)
  case '2':
    evenFinder(user_list)
  case '3':
    oddFinder(user_list)
  case '4':
    stringLengthFinder(user_list)
  case '5':
    empty(user_list)
  case _:
    print("Invalid Option")
```
